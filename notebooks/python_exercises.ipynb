```python
#list comprehension 
#instead of writing a loop, we use the loop inside a list to create another list

#Here we multiply each element in l by 2 and create a new list
new_l = [elt*2 for elt in l]
print(new_l)
```

```python
#functions
customer_data = [
    {'name': 'customer_1', 'id': 1, 'age': 100},
    {'name': 'customer_2', 'id': 2, 'age': 42},
    {'name': 'customer_3', 'id': 3, 'age': 25},
    {'name': 'customer_4', 'id': 4, 'age': 19},
]

def get_age_bucket(customer):
    customer_age = customer['age']
    if customer_age > 85:
        return '85+'
    elif customer_age > 50:
        return '50_85'
    elif customer_age > 30:
        return '30_50'
    else:
        return '0_30'

for customer in customer_data:
    print(customer['age'], get_age_bucket(customer))
```

```python
#2 Data Quality Checker
def email_scanner(emails):
    out={}
    out["valid_emails"]=[]
    out["invalid_emails"] = []
    for email in emails:
        if "@" not in email:
            out["invalid_emails"].append(email)
        elif email.rfind(".") < email.find("@"):
            out["invalid_emails"].append(email)
        elif email.strip().find("@") < 1:
            out["invalid_emails"].append(email)
        else:
            out["valid_emails"].append(email)

    #changes to remove duplicates
    out["invalid_emails"] = list(dict.fromkeys(out["invalid_emails"]))
    out["valid_emails"] = list(dict.fromkeys(out["valid_emails"]))

    return out


email_list = [
    "john.doe@company.com",
    "jane.smith@email.co.uk",
    "invalid-email",
    "bob@gmail.com",
    "alice.brown@company.com",
    "john.doe@company.com",  # duplicate
    "missing@domain",
    "test@example.org",
    "@nodomain.com",
    "jane.smith@email.co.uk",  # duplicate
    "valid.user@site.net",
    "no-at-symbol.com",
    "another@test.io"
]

print(email_scanner(email_list))
    

```

```python
#3 Sales Performance Tracker
class SalesRep:
    def __init__(self, some_value):
        self.some_value = some_value
    
    def total_sales(self):
        sales_out ={}
        for name in self:
            sales_out[name] = sum(self[name])
        return sales_out

    def average_sales(self):
        avg_out ={}
        for name in self:
            avg_out[name]= sum(self[name])/ len(self[name])
        return avg_out

    def hit_target_amount(self,amt):
        if total_sales(self) >= amt:
            return True
        else:
            return False



```
