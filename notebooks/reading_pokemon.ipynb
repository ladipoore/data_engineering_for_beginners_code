```python
#3
import sqlite3
import json

local_file = "pokemon_data.json"
database_file = "pokemon.db"

#Open a sqlite3 connection
conn = sqlite3.connect(database_file)
cursor = conn.cursor()

#creating a table
cursor.execute("create table pokebase(id, name, base_experience)")
conn.commit()
```

```python
res = cursor.execute("select name from sqlite_master")
res.fetchone()
```

```python

with open (local_file, "r") as f:
    values = json.load(f)
    # print(values.keys())
    for k,v in values.items():
        if k == 'id':
            id = v
        if k == 'name':
            name = v
        if k == 'base_experience':
            base = v
```

```python
cursor.execute("""
    insert into pokebase(id, name, base_experience)
    values (?,?,?)
""", (id, name, base))

conn.commit()
```

```python
res = cursor.execute("Select * from pokebase")
res.fetchall()
```
